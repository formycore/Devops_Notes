```python
# recursive walking through directories
from pathlib import Path
start_dir = Path("/home/maanya/Downloads/other_github/java_techie/")

for current_path, subdirectories, files in start_dir.rglob("*"):
    print("Current Path:", current_path)
    print("Subdirectories:", subdirectories)
    print("Files:", files)
    print("-" * 30)
```

```python
from pathlib import Path

start_dir = Path("/home/maanya/Downloads/other_github/java_techie/")

for path in start_dir.rglob("*"):
    if path.is_dir():
        print("Current Path:", path)
        subdirectories = [subdir for subdir in path.iterdir() if subdir.is_dir()]
        print("Subdirectories:", subdirectories)
        files = [file for file in path.iterdir() if file.is_file()]
        print("Files:", files)
        print("-" * 30)
```
